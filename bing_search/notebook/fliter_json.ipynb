```python
import os, sys, json, time, datetime, requests

json_dir = '/workspace/repo/cccrawl/webc/top65_urls'
json_file_list = os.listdir(json_dir)
json_file_list.sort()

for json_file in json_file_list:
	print('prcessing', json_file)
	contentUrl_set = set()
	new_json_data = []
	if not json_file.endswith('.json'):
		continue
	json_file_path = os.path.join(json_dir, json_file)
	with open(json_file_path, 'r') as f:
		json_data = json.load(f)
	print(json_file_path)
	for item in json_data:
		contentUrl = item['contentUrl']
		if contentUrl in contentUrl_set:
			continue
		contentUrl_set.add(contentUrl)
		new_json_data.append(item)
	with open(json_file_path, 'w') as f:
		json.dump(new_json_data, f, ensure_ascii=False, indent=4)
		print('write to', json_file_path)

	
```

```python
import os, sys, json, time, datetime, requests

json_dir = '/workspace/repo/cccrawl/data/top65/meta'
json_file_list = os.listdir(json_dir)
json_file_list.sort()

merged_json_data = []
for json_file in json_file_list:
	print('prcessing', json_file)
	if not json_file.endswith('.json'):
		continue
	json_file_path = os.path.join(json_dir, json_file)
	with open(json_file_path, 'r') as f:
		json_data = json.load(f)
	print(json_file_path)
	merged_json_data.extend(json_data)
with open('/workspace/repo/cccrawl/data/top65/meta.json', 'w') as f:
	json.dump(merged_json_data, f, ensure_ascii=False)
	print('write to', '/workspace/repo/cccrawl/data/top65/meta/_merged.json')
```
